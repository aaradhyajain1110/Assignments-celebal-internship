```python
def segment_signal(signal, window_sec=2, step_sec=1, Fs=200):
    window_size = window_sec * Fs
    step_size = step_sec * Fs
    segments = []

    for start in range(0, len(signal) - window_size + 1, step_size):
        end = start + window_size
        segment = signal[start:end]
        segments.append(segment)

    return segments
```

```python
import pywt
import numpy as np

def extract_band_power(signal, wavelet='db4', level=5):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    features = []
    for c in coeffs[1:]:  # skip approximation, use details
        features.append(np.log(np.sum(np.square(c))))  # energy
    return features
```
