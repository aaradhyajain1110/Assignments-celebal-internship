````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    try:
        query_lower = query.lower()
    # TODO: Implement routing logic

    # Example structure:

    # if "calculate" in query_lower:

    #     ...

    # elif "keywords" in query_lower:

    #     ...

    # else:

    #     ...

        if "calculate" in query_lower:
            # Pull out everything after the word "calculate"
            expression = query_lower.split("calculate", 1)[1].strip()
            if not expression:
                return {"type": "error", "result": "No expression provided to calculate"}
            result = calculator(expression)
            if result == "Error in calculation":
                return {"type": "error", "result": result}
            return {"type": "calculation", "result": result}

        elif "keywords" in query_lower:
            # Grab the text after "keywords from" if present, else whole query
            if "keywords from" in query_lower:
                text = query.split("keywords from", 1)[1].strip()
            else:
                text = query
            if not text:
                return {"type": "error", "result": "No text provided for keyword extraction"}
            result = extract_keywords(text)
            return {"type": "keywords", "result": result}

        else:
            return {"type": "general", "result": query}

    except Exception as e:
        return {"type": "error", "result": f"Something went wrong: {str(e)}"}
````

````python
# 🤖 AGENT FUNCTION (IMPROVED ROUTING + LOGGING)

query_log = []  # keeps a history of every query + response

def agent(query: str):
    try:
        query_lower = query.lower()

        # Improved routing: catch more trigger words for calculation
        calc_triggers = ["calculate", "calc", "solve", "compute"]
        keyword_triggers = ["keywords", "extract keywords", "find keywords"]

        if any(trigger in query_lower for trigger in calc_triggers):
            expression = query_lower
            for trigger in calc_triggers:
                expression = expression.replace(trigger, "")
            expression = expression.strip()

            if not expression:
                response = {"type": "error", "result": "No expression provided to calculate"}
            else:
                result = calculator(expression)
                if result == "Error in calculation":
                    response = {"type": "error", "result": result}
                else:
                    response = {"type": "calculation", "result": result}

        elif any(trigger in query_lower for trigger in keyword_triggers):
            if "keywords from" in query_lower:
                text = query.split("keywords from", 1)[1].strip()
            else:
                text = query
            if not text:
                response = {"type": "error", "result": "No text provided for keyword extraction"}
            else:
                result = extract_keywords(text)
                response = {"type": "keywords", "result": result}

        else:
            response = {"type": "general", "result": query}

    except Exception as e:
        response = {"type": "error", "result": f"Something went wrong: {str(e)}"}

    # 📝 Logging: save every query + response pair
    query_log.append({"query": query, "response": response})

    return response
````

````python
# 🛠️ TOOL 3: Word Counter

def word_counter(text: str) -> dict:
    """Count words, characters, and sentences in text."""
    try:
        words = text.split()
        word_count = len(words)
        char_count = len(text.replace(" ", ""))
        sentence_count = text.count(".") + text.count("!") + text.count("?")
        sentence_count = max(sentence_count, 1)  # at least 1 if there's text
        return {
            "words": word_count,
            "characters": char_count,
            "sentences": sentence_count
        }
    except Exception:
        return {"error": "Could not count words"}
````

````python
# 🤖 AGENT FUNCTION (IMPROVED ROUTING + LOGGING + 3 TOOLS)

query_log = []  # keeps a history of every query + response

def agent(query: str):
    try:
        query_lower = query.lower()

        calc_triggers = ["calculate", "calc", "solve", "compute"]
        keyword_triggers = ["keywords", "extract keywords", "find keywords"]
        count_triggers = ["count words", "word count", "count text"]

        if any(trigger in query_lower for trigger in calc_triggers):
            expression = query_lower
            for trigger in calc_triggers:
                expression = expression.replace(trigger, "")
            expression = expression.strip()

            if not expression:
                response = {"type": "error", "result": "No expression provided to calculate"}
            else:
                result = calculator(expression)
                if result == "Error in calculation":
                    response = {"type": "error", "result": result}
                else:
                    response = {"type": "calculation", "result": result}

        elif any(trigger in query_lower for trigger in keyword_triggers):
            if "keywords from" in query_lower:
                text = query.split("keywords from", 1)[1].strip()
            else:
                text = query
            if not text:
                response = {"type": "error", "result": "No text provided for keyword extraction"}
            else:
                result = extract_keywords(text)
                response = {"type": "keywords", "result": result}

        elif any(trigger in query_lower for trigger in count_triggers):
            text = query
            for trigger in count_triggers:
                text = text.lower().replace(trigger, "")
            text = text.strip()
            if not text:
                response = {"type": "error", "result": "No text provided to count"}
            else:
                result = word_counter(text)
                response = {"type": "word_count", "result": result}

        else:
            response = {"type": "general", "result": query}

    except Exception as e:
        response = {"type": "error", "result": f"Something went wrong: {str(e)}"}

    query_log.append({"query": query, "response": response})

    return response
````
